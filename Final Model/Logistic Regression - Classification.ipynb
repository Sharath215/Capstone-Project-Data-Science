```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
dataset = pd.read_csv("sales_Price.csv")  # reading the dataset 
```

```python
dataset.head()
```

```python
dataset1 = dataset.drop(['Unnamed: 0','zipcode','lat','long'], axis = 1, inplace = True)
```

```python
dataset.head(5)
```

```python
dataset['Rating'] = ['1' if x == 'Excellent' else '0' for x in dataset['Rating']]
```

```python
dataset.head()
```

```python
dependent = dataset[['Rating']]
independent = dataset.drop('Rating', axis = 1)
```

```python
independent.shape
```

```python
dependent
```

```python
# split the dataset into training and test

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
dependent
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(penalty = 'l2', solver = 'newton-cg',random_state = 0)
classifier.fit(X_train, y_train)                           # Logistic regression classification
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, y_pred)
```

```python
print(clf_report)
```
