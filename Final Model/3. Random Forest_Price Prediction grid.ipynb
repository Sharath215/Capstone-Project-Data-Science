```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from UnivariateFunctions import Univariate as Unv
```

```python
dataset = pd.read_csv("sales_Price.csv")
dataset 
```

```python
dataset = dataset.drop(['Unnamed: 0','zipcode','lat','long'], axis = 1)
```

```python
dataset = pd.get_dummies(dataset, dtype=int)
```

```python
dataset.fillna(0 , inplace = True) # replacing with NA
dataset.isnull().sum()
```

```python
dataset
```

```python
dataset = dataset.drop(['Rating_Good'],axis = 1)
```

```python
#changing both values to int

dataset['bathrooms'] = dataset['bathrooms'].apply(np.int64)
dataset['price'] = dataset['price'].apply(np.int64)
```

```python
# classification is the output variable, so storing it in y, x is input

independent = dataset.iloc[:,1:].values
dependent =dataset.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(independent,dependent,test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor  # importing Support vector machine model.

param_grid = {'criterion':['absolute_error', 'squared_error', 'poisson', 'friedman_mse'],'max_features':['auto','sqrt','log2'],'n_estimators':[10,100]}

grid = GridSearchCV(RandomForestRegressor(),param_grid, refit = True, verbose = 3, n_jobs=-1)


grid.fit(X_train,y_train)
```

```python
# grid_predictions = grid.predict(X_test)
# from sklearn.metrics import r2_score
# r_score = r2_score(y_test, grid_predictions)

re = grid.cv_results_

print("the R score value for best parameter {}:".format(grid.best_params_))
```

```python
table = pd.DataFrame.from_dict(re)
table
```
