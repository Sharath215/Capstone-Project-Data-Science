```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from UnivariateFunctions import Univariate as Unv
```

```python
dataset = pd.read_csv("sales_Price.csv")
dataset 
```

```python
dataset = dataset.drop(['Unnamed: 0','zipcode','lat','long'], axis = 1)
```

```python
dataset = pd.get_dummies(dataset, dtype=int)
```

```python
dataset.fillna(0 , inplace = True) # replacing with NA
dataset.isnull().sum()
```

```python
dataset
```

```python
dataset = dataset.drop(['Rating_Good'],axis = 1)
```

```python
#changing both values to int

dataset['bathrooms'] = dataset['bathrooms'].apply(np.int64)
dataset['price'] = dataset['price'].apply(np.int64)
```

```python
# classification is the output variable, so storing it in y, x is input

independent = dataset.iloc[:,1:].values
dependent =dataset.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(independent,dependent,test_size = 0.30, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(criterion = 'absolute_error',max_features='sqrt',n_estimators=100,random_state=0)
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred)
r_score
```
